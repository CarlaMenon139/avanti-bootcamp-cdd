```python
import pandas as pd

# Carregar o arquivo CSV para análise
file_path = r'C:\Users\menon\OneDrive\Documentos\Menon\CarlaMenon\diamonsts.csv'


data = pd.read_csv(file_path)

# Obter informações básicas dos dados
info = data.info()
missing_values = data.isnull().sum()
data_head = data.head()

info, missing_values, data_head


# Recarregar o arquivo usando o delimitador correto (;)
data = pd.read_csv(file_path, delimiter=';')

# Obter informações básicas dos dados
info = data.info()
missing_values = data.isnull().sum()
data_head = data.head()

info, missing_values, data_head

import seaborn as sns
import matplotlib.pyplot as plt

# Configuração do estilo para visualizações
sns.set(style="whitegrid")

# Relação entre o peso (carat) e o preço do diamante
plt.figure(figsize=(10, 6))
sns.scatterplot(x='carat', y='price', data=data, alpha=0.5)
plt.title('Relação entre Carat e Preço')
plt.xlabel('Carat (Peso)')
plt.ylabel('Preço (USD)')
plt.show()

# Preço médio por qualidade de corte
plt.figure(figsize=(10, 6))
sns.boxplot(x='cut', y='price', data=data)
plt.title('Preço por Qualidade de Corte')
plt.xlabel('Qualidade de Corte')
plt.ylabel('Preço (USD)')
plt.show()

# Preço médio por clareza do diamante
plt.figure(figsize=(10, 6))
sns.boxplot(x='clarity', y='price', data=data, order=data['clarity'].value_counts().index)
plt.title('Preço por Clareza')
plt.xlabel('Clareza')
plt.ylabel('Preço (USD)')
plt.show()

#Salvar o código em arquivo computador
data.to_csv('C:\\Users\\menon\\OneDrive\\Documentos\\Menon\\CarlaMenon\\diamonts-analise_exploratoria.csv', index=False)



```
