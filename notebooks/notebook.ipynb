```python
import pandas as pd

# Carregar o arquivo CSV para anlise
file_path = r'C:\Users\menon\OneDrive\Documentos\Menon\CarlaMenon\diamonsts.csv'


data = pd.read_csv(file_path)

# Obter informaes bsicas dos dados
info = data.info()
missing_values = data.isnull().sum()
data_head = data.head()

info, missing_values, data_head


# Recarregar o arquivo usando o delimitador correto (;)
data = pd.read_csv(file_path, delimiter=';')

# Obter informaes bsicas dos dados
info = data.info()
missing_values = data.isnull().sum()
data_head = data.head()

info, missing_values, data_head

import seaborn as sns
import matplotlib.pyplot as plt

# Configurao do estilo para visualizaes
sns.set(style="whitegrid")

# Relao entre o peso (carat) e o preo do diamante
plt.figure(figsize=(10, 6))
sns.scatterplot(x='carat', y='price', data=data, alpha=0.5)
plt.title('Relao entre Carat e Preo')
plt.xlabel('Carat (Peso)')
plt.ylabel('Preo (USD)')
plt.show()

# Preo mdio por qualidade de corte
plt.figure(figsize=(10, 6))
sns.boxplot(x='cut', y='price', data=data)
plt.title('Preo por Qualidade de Corte')
plt.xlabel('Qualidade de Corte')
plt.ylabel('Preo (USD)')
plt.show()

# Preo mdio por clareza do diamante
plt.figure(figsize=(10, 6))
sns.boxplot(x='clarity', y='price', data=data, order=data['clarity'].value_counts().index)
plt.title('Preo por Clareza')
plt.xlabel('Clareza')
plt.ylabel('Preo (USD)')
plt.show()

#Salvar o cdigo em arquivo computador
data.to_csv('C:\\Users\\menon\\OneDrive\\Documentos\\Menon\\CarlaMenon\\diamonts-analise_exploratoria.csv', index=False)



```
